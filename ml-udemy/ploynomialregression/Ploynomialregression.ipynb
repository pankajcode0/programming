# PloynomialLinearRegression

# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Data

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
print(dataset)
```

```python
print(x)
print("------------------------------")
print(y)
```

# Applying Linear Regression 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
```

# Visualising Linear Regression results

```python
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x))
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Example: Predicting salary of 5.5 level

```python
salary = regressor.predict([[6.5]])
print(salary)
```

# Applying Polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

# Visualising Polynomial Linear Regression

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly))
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# smoother curve ploting

```python
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting Salary using polynomial regressor model

```python
salary2 = lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
print(salary2)
```
